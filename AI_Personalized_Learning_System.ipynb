# üß† Core AI Implementation

The following code blocks represent the **core artificial intelligence and personalization logic**
used in the *AI Personalized Learning System*.  
These implementations are **directly reflected in the Streamlit application** and demonstrate
adaptive learning, learner profiling, recommendation generation, and performance analytics.


```python
# Learner Profiling Logic

def build_learner_profile(accuracy, avg_response_time, total_attempts):
    """
    Builds a learner profile based on performance metrics
    """

    # Engagement
    if total_attempts >= 20:
        engagement = "High"
    elif total_attempts >= 10:
        engagement = "Medium"
    else:
        engagement = "Low"

    # Pace
    if avg_response_time < 10:
        pace = "Fast"
    elif avg_response_time < 20:
        pace = "Moderate"
    else:
        pace = "Slow"

    # Confidence
    if accuracy >= 75:
        confidence = "High"
    elif accuracy >= 50:
        confidence = "Medium"
    else:
        confidence = "Low"

    return {
        "engagement": engagement,
        "pace": pace,
        "confidence": confidence
    }
```

```python
# Adaptive Difficulty Engine

def adaptive_difficulty_engine(score, total_questions):
    """
    Dynamically adjusts quiz difficulty based on learner accuracy
    """

    accuracy = (score / max(1, total_questions)) * 100

    if accuracy >= 80:
        difficulty = "Hard"
    elif accuracy >= 50:
        difficulty = "Medium"
    else:
        difficulty = "Easy"

    return difficulty, accuracy
```

```python
# Quiz Evaluation & Feedback System

def evaluate_answer(user_answer, correct_answer, response_time):
    """
    Evaluates answer correctness and response behavior
    """

    is_correct = user_answer == correct_answer

    if is_correct and response_time < 10:
        feedback = "Excellent speed and accuracy!"
    elif is_correct:
        feedback = "Correct answer, keep improving speed."
    else:
        feedback = "Incorrect. Review the concept and try again."

    return is_correct, feedback
```

```python
# Recommendation Engine
def generate_recommendations(topic_accuracy):
    """
    Generates personalized learning recommendations
    """

    recommendations = []

    for topic, acc in topic_accuracy.items():
        if acc < 60:
            recommendations.append(
                f"Revise basics of {topic} and practice easy questions."
            )
        elif acc < 80:
            recommendations.append(
                f"Practice medium-level questions in {topic}."
            )
        else:
            recommendations.append(
                f"Advance to harder concepts in {topic}."
            )

    return recommendations
```

```python
# Overall Performance Score Calculation

def compute_overall_score(accuracy, avg_time, consistency):
    """
    Computes weighted overall performance score
    """

    score = (
        (accuracy * 0.6) +
        ((100 - avg_time) * 0.25) +
        (consistency * 0.15)
    )

    return round(score, 2)
```

```python
# Learning History Logging (DATA PERSISTENCE)

from datetime import datetime

def log_learning_activity(student_id, topic, accuracy):
    """
    Logs learner activity
    """

    return {
        "student_id": student_id,
        "topic": topic,
        "accuracy": accuracy,
        "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    }
```

```python
# Badge & Level Assignment (GAMIFICATION)

def assign_badge(accuracy):
    """
    Assigns badge based on performance
    """

    if accuracy >= 90:
        return "üèÜ Master"
    elif accuracy >= 75:
        return "üöÄ Advanced"
    elif accuracy >= 50:
        return "üå± Beginner"
    else:
        return "üìò Needs Practice"
```

```python
# End to End Flow

# Simulated learner session

score = 7
total_questions = 10
avg_response_time = 12
total_attempts = 18
topic_accuracy = {
    "Mathematics": 55,
    "Programming": 82,
    "Science": 65
}

difficulty, accuracy = adaptive_difficulty_engine(score, total_questions)
profile = build_learner_profile(accuracy, avg_response_time, total_attempts)
recommendations = generate_recommendations(topic_accuracy)
badge = assign_badge(accuracy)

difficulty, profile, recommendations, badge
```

1Ô∏è‚É£ Explainable Adaptive Difficulty (WHY difficulty changed)

```python
def explain_difficulty_change(previous_difficulty, new_difficulty, accuracy):
    """
    Provides human-readable explanation
    for adaptive difficulty changes
    """

    if previous_difficulty == new_difficulty:
        return (
            f"Difficulty maintained at {new_difficulty} "
            f"due to stable accuracy ({accuracy:.1f}%)."
        )

    if new_difficulty == "Hard":
        return (
            f"Difficulty increased to Hard because accuracy "
            f"reached {accuracy:.1f}%, indicating strong mastery."
        )

    if new_difficulty == "Easy":
        return (
            f"Difficulty reduced to Easy because accuracy "
            f"dropped to {accuracy:.1f}%, ensuring better learning support."
        )

    return "Difficulty adjusted based on learner performance."
```

2Ô∏è‚É£ Learning Pace Classification

```python
def classify_learning_pace(avg_response_time):
    """
    Classifies learner pace based on response time
    """

    if avg_response_time < 8:
        return "Fast Learner"
    elif avg_response_time < 15:
        return "Moderate Learner"
    else:
        return "Slow Learner"
```

3Ô∏è‚É£ Knowledge Retention Score (Advanced Metric)

```python
def compute_retention_score(accuracy, attempts):
    """
    Estimates knowledge retention using
    performance consistency
    """

    if attempts == 0:
        return 0

    retention = accuracy * (min(attempts, 20) / 20)
    return round(retention, 2)
```

4Ô∏è‚É£ Weak Topic Detector (Used in Profile Page)

```python
def detect_weak_topics(topic_accuracy, threshold=60):
    """
    Identifies weak learning areas
    """

    weak_topics = []

    for topic, acc in topic_accuracy.items():
        if acc < threshold:
            weak_topics.append({
                "topic": topic,
                "accuracy": acc,
                "action": "Revise fundamentals and practice more questions"
            })

    return weak_topics
```

5Ô∏è‚É£ Smart Resource Recommendation Layer

```python
def recommend_learning_resources(topic, accuracy):
    """
    Recommends external resources based on performance
    """

    base_resources = {
        "Mathematics": "https://www.khanacademy.org/math",
        "Programming": "https://www.geeksforgeeks.org",
        "Science": "https://www.khanacademy.org/science",
        "Languages": "https://www.grammarly.com/blog"
    }

    if accuracy < 50:
        level = "Beginner revision"
    elif accuracy < 75:
        level = "Intermediate practice"
    else:
        level = "Advanced concepts"

    return {
        "topic": topic,
        "recommended_level": level,
        "resource_link": base_resources.get(topic, "https://www.wikipedia.org")
    }
```


6Ô∏è‚É£ Confidence Growth Model (Behavioral AI)

```python
def confidence_growth(previous_confidence, current_accuracy):
    """
    Models confidence progression over time
    """

    if current_accuracy >= 80:
        return "High"
    elif current_accuracy >= 60:
        return "Medium"
    else:
        return previous_confidence
```

7Ô∏è‚É£ AI-Based Learner Summary Generator

```python
def generate_ai_summary(profile, accuracy, badge):
    """
    Generates a human-readable learner summary
    """

    return (
        f"The learner demonstrates {profile['engagement']} engagement "
        f"with a {profile['pace']} learning pace. "
        f"Confidence level is {profile['confidence']}, "
        f"and the current achievement badge is {badge}. "
        f"Overall accuracy stands at {accuracy:.1f}%."
    )
```

8Ô∏è‚É£ End-to-End Evaluation Simulation

```python
# ---------- FULL PIPELINE DEMO ----------

pace_label = classify_learning_pace(avg_response_time)
retention_score = compute_retention_score(accuracy, total_attempts)
weak_areas = detect_weak_topics(topic_accuracy)

difficulty_explanation = explain_difficulty_change(
    previous_difficulty="Medium",
    new_difficulty=difficulty,
    accuracy=accuracy
)

resource_plan = recommend_learning_resources("Mathematics", topic_accuracy["Mathematics"])

ai_summary = generate_ai_summary(profile, accuracy, badge)

{
    "Learner Pace": pace_label,
    "Retention Score": retention_score,
    "Weak Areas": weak_areas,
    "Difficulty Explanation": difficulty_explanation,
    "Recommended Resource": resource_plan,
    "AI Summary": ai_summary
}
```
