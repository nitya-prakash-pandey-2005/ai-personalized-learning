```python
# Learner Profiling Logic

def build_learner_profile(accuracy, avg_response_time, total_attempts):
    """
    Builds a learner profile based on performance metrics
    """

    # Engagement
    if total_attempts >= 20:
        engagement = "High"
    elif total_attempts >= 10:
        engagement = "Medium"
    else:
        engagement = "Low"

    # Pace
    if avg_response_time < 10:
        pace = "Fast"
    elif avg_response_time < 20:
        pace = "Moderate"
    else:
        pace = "Slow"

    # Confidence
    if accuracy >= 75:
        confidence = "High"
    elif accuracy >= 50:
        confidence = "Medium"
    else:
        confidence = "Low"

    return {
        "engagement": engagement,
        "pace": pace,
        "confidence": confidence
    }
```

```python
# Adaptive Difficulty Engine

def adaptive_difficulty_engine(score, total_questions):
    """
    Dynamically adjusts quiz difficulty based on learner accuracy
    """

    accuracy = (score / max(1, total_questions)) * 100

    if accuracy >= 80:
        difficulty = "Hard"
    elif accuracy >= 50:
        difficulty = "Medium"
    else:
        difficulty = "Easy"

    return difficulty, accuracy
```

```python
# Quiz Evaluation & Feedback System

def evaluate_answer(user_answer, correct_answer, response_time):
    """
    Evaluates answer correctness and response behavior
    """

    is_correct = user_answer == correct_answer

    if is_correct and response_time < 10:
        feedback = "Excellent speed and accuracy!"
    elif is_correct:
        feedback = "Correct answer, keep improving speed."
    else:
        feedback = "Incorrect. Review the concept and try again."

    return is_correct, feedback
```

```python
# Recommendation Engine
def generate_recommendations(topic_accuracy):
    """
    Generates personalized learning recommendations
    """

    recommendations = []

    for topic, acc in topic_accuracy.items():
        if acc < 60:
            recommendations.append(
                f"Revise basics of {topic} and practice easy questions."
            )
        elif acc < 80:
            recommendations.append(
                f"Practice medium-level questions in {topic}."
            )
        else:
            recommendations.append(
                f"Advance to harder concepts in {topic}."
            )

    return recommendations
```

```python
# Overall Performance Score Calculation

def compute_overall_score(accuracy, avg_time, consistency):
    """
    Computes weighted overall performance score
    """

    score = (
        (accuracy * 0.6) +
        ((100 - avg_time) * 0.25) +
        (consistency * 0.15)
    )

    return round(score, 2)
```

```python
# Learning History Logging (DATA PERSISTENCE)

from datetime import datetime

def log_learning_activity(student_id, topic, accuracy):
    """
    Logs learner activity
    """

    return {
        "student_id": student_id,
        "topic": topic,
        "accuracy": accuracy,
        "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    }
```

```python
# Badge & Level Assignment (GAMIFICATION)

def assign_badge(accuracy):
    """
    Assigns badge based on performance
    """

    if accuracy >= 90:
        return "ðŸ† Master"
    elif accuracy >= 75:
        return "ðŸš€ Advanced"
    elif accuracy >= 50:
        return "ðŸŒ± Beginner"
    else:
        return "ðŸ“˜ Needs Practice"
```

```python
# End to End Flow

# Simulated learner session

score = 7
total_questions = 10
avg_response_time = 12
total_attempts = 18
topic_accuracy = {
    "Mathematics": 55,
    "Programming": 82,
    "Science": 65
}

difficulty, accuracy = adaptive_difficulty_engine(score, total_questions)
profile = build_learner_profile(accuracy, avg_response_time, total_attempts)
recommendations = generate_recommendations(topic_accuracy)
badge = assign_badge(accuracy)

difficulty, profile, recommendations, badge
```
